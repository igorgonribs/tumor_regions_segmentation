```python
## Training from zero (1)
import pandas as pd 
from IPython.core.display import HTML

training_dir = "../../datasets/BONE_CHANNELS/training"
csv_file_path = "{}/bones_training_accuracy_loss_1.csv".format(training_dir)
df = pd.read_csv(csv_file_path) 

train = df[df['phase'] == 'train']
valid = df[df['phase'] == 'test']

train_best_accuracy = train[train['accuracy']==train['accuracy'].max()]
display(HTML(train_best_accuracy.to_html()))

valid_best_accuracy = valid[valid['accuracy']==valid['accuracy'].max()]
display(HTML(valid_best_accuracy.to_html()))

df_loss_lines_1 = pd.DataFrame({
    'loss train (random)': train['loss'].to_numpy(),
    'loss valid (random)': valid['loss'].to_numpy()},
    index = train['epoch'])

df_accuracy_lines_1 = pd.DataFrame({
    'acc train (random)': train['accuracy'].to_numpy(),
    'acc valid (random)': valid['accuracy'].to_numpy()}, 
    index = train['epoch'])
```

```python
df_accuracy_lines_1['acc train (random)']=df_accuracy_lines_1['acc train (random)'].astype(float)
df_accuracy_lines_1['acc valid (random)']=df_accuracy_lines_1['acc valid (random)'].astype(float)
accuracy_lines_1 = df_accuracy_lines_1['acc train (random)'].plot.line(title='Accuracy Train', color='red')
```

```python
df_loss_lines_1['loss train (random)']=df_loss_lines_1['loss train (random)'].astype(float)
df_loss_lines_1['loss valid (random)']=df_loss_lines_1['loss valid (random)'].astype(float)
loss_lines_1 = df_loss_lines_1['loss train (random)'].plot.line(title='Loss train')
```

```python
## Transfer learning

import pandas as pd 
from IPython.core.display import HTML

training_dir = "../../datasets/BONE_CHANNELS/training"
csv_file_path = "{}/bones_training_accuracy_loss_transfer_learning.csv".format(training_dir)
df = pd.read_csv(csv_file_path) 

train = df[df['phase'] == 'train']
valid = df[df['phase'] == 'test']

train_best_accuracy = train[train['accuracy']==train['accuracy'].max()]
display(HTML(train_best_accuracy.to_html()))

valid_best_accuracy = valid[valid['accuracy']==valid['accuracy'].max()]
display(HTML(valid_best_accuracy.to_html()))

df_loss_lines_TL = pd.DataFrame({
    'loss train (random)': train['loss'].to_numpy(),
    'loss valid (random)': valid['loss'].to_numpy()},
    index = train['epoch'])

df_accuracy_lines_TL = pd.DataFrame({
    'acc train (random)': train['accuracy'].to_numpy(),
    'acc valid (random)': valid['accuracy'].to_numpy()}, 
    index = train['epoch'])

```

```python
df_accuracy_lines_TL['acc train (random)']=df_accuracy_lines_TL['acc train (random)'].astype(float)
df_accuracy_lines_TL['acc valid (random)']=df_accuracy_lines_TL['acc valid (random)'].astype(float)
accuracy_lines_TL = df_accuracy_lines_TL['acc train (random)'].plot.line(title='Accuracy Train', color='red')
```

```python
df_loss_lines_TL['loss train (random)']=df_loss_lines_TL['loss train (random)'].astype(float)
df_loss_lines_TL['loss valid (random)']=df_loss_lines_TL['loss valid (random)'].astype(float)
loss_lines_TL = df_loss_lines_TL['loss train (random)'].plot.line(title='Loss train')
```

```python
# importing package 
import matplotlib.pyplot as plt 
  
  
# plot lines 
plt.plot(df_loss_lines_1['loss train (random)'], label = "Treinamento do zero (conjunto de treinamento)")
plt.plot(df_loss_lines_TL['loss train (random)'], label = "Aprendizado transferido (conjunto de treinamento)")

plt.plot(df_loss_lines_1['loss valid (random)'], label = "Treinamento do zero (conjunto de validação)")
plt.plot(df_loss_lines_TL['loss valid (random)'], label = "Aprendizado transferido (conjunto de validação)")

#plt.legend() 
plt.legend(loc='center left', bbox_to_anchor=(1, 0.85))
plt.title('Erro x épocas')
plt.show()
```

```python
# importing package 
import matplotlib.pyplot as plt 
  
  
# plot lines 
plt.plot(df_accuracy_lines_1['acc train (random)'], label = "Treinamento do zero (conjunto de treinamento)")
plt.plot(df_accuracy_lines_TL['acc train (random)'], label = "Aprendizado transferido (conjunto de treinamento)")

plt.plot(df_accuracy_lines_1['acc valid (random)'], label = "Treinamento do zero (conjunto de validação)")
plt.plot(df_accuracy_lines_TL['acc valid (random)'], label = "Aprendizado transferido (conjunto de validação)")

#plt.legend() 
plt.legend(loc='center left', bbox_to_anchor=(1, 0.85))
plt.title('Acurácia x épocas')
plt.show()
```
